```python
def check_to_fill(data, x, y):
    if data[x-1][y-1] == '#' or data[x-1][y] == '#' or data[x-1][y+1] == '#':
        return False
    elif data[x][y-1] == '#' or data[x][y+1] == '#':
        return False
    elif data[x+1][y-1] == '#' or data[x+1][y] == '#' or data[x+1][y+1] == '#':
        return False
    else:
        return True
    
def check_to_vacate(data, x, y):
    
    counter = 0
    
    if data[x-1][y-1] == '#':
        counter += 1
    if data[x-1][y] == '#':
        counter += 1
    if data[x-1][y+1] == '#':
        counter += 1
    if data[x][y-1] == '#':
        counter += 1
    if data[x][y+1] == '#':
        counter += 1
    if data[x+1][y-1] == '#':
        counter += 1
    if data[x+1][y] == '#':
        counter +=1
    if data[x+1][y+1] == '#':
        counter += 1
    
    if counter >= 4:
        return True
    
    return False

def get_occupied_seats(data):
    
    repeat = True
    
    num_rows = len(data)
    num_cols = len(data[0])
    
    iterations = 0
    
    while repeat:
        
        iterations += 1
        data_copy = data.copy()
        
        for x in range(num_rows):
            for y in range(num_cols):
                
                edge = False
                
                if x == 0:
                    edge = True
                    
                    if y == 0 :
                        if data_copy[x][y] == 'L' and data_copy[x+1][y] != '#' and data_copy[x][y+1] != '#' and data_copy[x+1][y+1] != '#':
                            data[x] = data[x][:y] + '#' + data[x][y+1:]
                    elif y == num_cols - 1:
                        if data_copy[x][y] == 'L' and data_copy[x+1][y] != '#' and data_copy[x][y-1] != '#' and data_copy[x+1][y-1] != '#':
                            data[x] = data[x][:y] + '#' + data[x][y+1:]
                    else:
                        if data_copy[x][y] == 'L' and data_copy[x+1][y] != '#' and data_copy[x][y+1] != '#' and data_copy[x+1][y+1] != '#' and data_copy[x][y-1] != '#' and data_copy[x+1][y-1] != '#':
                            data[x] = data[x][:y] + '#' + data[x][y+1:]
                        
                        if data_copy[x][y] == '#':
                            counter = 0
    
                            if data_copy[x][y-1] == '#':
                                counter += 1
                            if data_copy[x][y+1] == '#':
                                counter += 1
                            if data_copy[x+1][y-1] == '#':
                                counter += 1
                            if data_copy[x+1][y] == '#':
                                counter +=1
                            if data_copy[x+1][y+1] == '#':
                                counter += 1
                            
                            if counter >= 4:
                                data[x] = data[x][:y] + 'L' + data[x][y+1:]
                        
                elif x == num_rows - 1:
                    edge = True
                    
                    if y == 0 :
                        if data_copy[x][y] == 'L' and data_copy[x-1][y] != '#' and data_copy[x][y+1] != '#' and data_copy[x-1][y+1] != '#':
                            data[x] = data[x][:y] + '#' + data[x][y+1:]
                    elif y == num_cols - 1:
                        if data_copy[x][y] == 'L' and data_copy[x-1][y] != '#' and data_copy[x][y-1] != '#' and data_copy[x-1][y-1] != '#':
                            data[x] = data[x][:y] + '#' + data[x][y+1:]
                    else:
                        if data_copy[x][y] == 'L' and data_copy[x-1][y] != '#' and data_copy[x][y+1] != '#' and data_copy[x-1][y+1] != '#' and data_copy[x][y-1] != '#' and data_copy[x-1][y-1] != '#':
                            data[x] = data[x][:y] + '#' + data[x][y+1:]
                        
                        if data_copy[x][y] == '#':
                            counter = 0
    
                            if data_copy[x][y-1] == '#':
                                counter += 1
                            if data_copy[x][y+1] == '#':
                                counter += 1
                            if data_copy[x-1][y-1] == '#':
                                counter += 1
                            if data_copy[x-1][y] == '#':
                                counter +=1
                            if data_copy[x-1][y+1] == '#':
                                counter += 1
                            
                            if counter >= 4:
                                data[x] = data[x][:y] + 'L' + data[x][y+1:]
                
                if y == 0 and x != 0 and x != num_rows - 1:
                    edge = True
                    
                    if data_copy[x][y] == 'L' and data_copy[x-1][y] != '#' and data_copy[x][y+1] != '#' and data_copy[x-1][y+1] != '#' and data_copy[x+1][y+1] != '#' and data_copy[x+1][y] != '#':
                        data[x] = data[x][:y] + '#' + data[x][y+1:]
                    
                    if data_copy[x][y] == '#':
                        counter = 0

                        if data_copy[x+1][y] == '#':
                            counter += 1
                        if data_copy[x+1][y+1] == '#':
                            counter += 1
                        if data_copy[x-1][y+1] == '#':
                            counter += 1
                        if data_copy[x-1][y] == '#':
                            counter +=1
                        if data_copy[x][y+1] == '#':
                            counter += 1
                        
                        if counter >= 4:
                            data[x] = data[x][:y] + 'L' + data[x][y+1:]
                    
                if y == num_cols - 1 and x != 0 and x != num_rows - 1:
                    edge = True
                    
                    if data_copy[x][y] == 'L' and data_copy[x-1][y] != '#' and data_copy[x][y-1] != '#' and data_copy[x-1][y-1] != '#' and data_copy[x+1][y-1] != '#' and data_copy[x+1][y] != '#':
                        data[x] = data[x][:y] + '#' + data[x][y+1:]
                    
                    if data_copy[x][y] == '#':
                        counter = 0

                        if data_copy[x+1][y] == '#':
                            counter += 1
                        if data_copy[x+1][y-1] == '#':
                            counter += 1
                        if data_copy[x-1][y-1] == '#':
                            counter += 1
                        if data_copy[x-1][y] == '#':
                            counter +=1
                        if data_copy[x][y-1] == '#':
                            counter += 1
                        
                        if counter >= 4:
                            data[x] = data[x][:y] + 'L' + data[x][y+1:]
                    
                if not edge:
                    if data_copy[x][y] == 'L':
                        
                        if check_to_fill(data_copy, x, y):
                            data[x] = data[x][:y] + '#' + data[x][y+1:]
                            
                    elif data_copy[x][y] == '#':
                        
                        if check_to_vacate(data_copy, x, y):
                            data[x] = data[x][:y] + 'L' + data[x][y+1:]
        
        if data_copy == data:
            repeat = False
    
    seat_count = 0
    
    for row in data:
        seat_count += row.count('#')
    
    with open('output.txt', 'w') as output:
        for item in data:
            output.write("%s\n" % item)
    
    return seat_count
```

```python
def check_to_fill2(data, x, y):
    
    num_rows = len(data)
    num_cols = len(data[0])
    
    i = 1
    while x - i >= 0 and y - i >= 0: 
        if data[x-i][y-i] == '#':
            return False
        elif data[x-i][y-i] == 'L':
            i = num_rows + num_cols
        i += 1
    
    i = 1
    while x - i >= 0:
        if data[x-i][y] == '#':
            return False
        elif data[x-i][y] == 'L':
            i = num_rows + num_cols
        i += 1
        
    i = 1
    while x - i >= 0 and y + i <= num_cols - 1:
        if data[x-i][y+i] == '#':
            return False
        elif data[x-i][y+i] == 'L':
            i = num_rows + num_cols
        i += 1
    
    i = 1
    while y - i >= 0:
        if data[x][y-i] == '#':
            return False
        elif data[x][y-i] == 'L':
            i = num_rows + num_cols
        i += 1
    
    i = 1
    while y + i <= num_cols - 1:
        if data[x][y+i] == '#':
            return False
        elif data[x][y+i] == 'L':
            i = num_rows + num_cols
        i += 1
    
    i = 1
    while x + i <= num_rows - 1 and y - i >=0:
        if data[x+i][y-i] == '#':
            return False
        elif data[x+i][y-i] == 'L':
            i = num_rows + num_cols
        i += 1
    
    i = 1
    while x + i <= num_rows - 1:
        if data[x+i][y] == '#':
            return False
        elif data[x+i][y] == 'L':
            i = num_rows + num_cols
        i += 1
    
    i = 1
    while x + i <= num_rows - 1 and y + i <= num_cols - 1:
        if data[x+i][y+i] == '#':
            return False
        elif data[x+i][y+i] == 'L':
            i = num_rows + num_cols
        i += 1
        
    return True

def check_to_vacate2(data, x, y):
    
    counter = 0
    
    num_rows = len(data)
    num_cols = len(data[0])
    
    i = 1
    while x - i >= 0 and y - i >= 0: 
        if data[x-i][y-i] == '#':
            counter += 1
            i = num_rows + num_cols
        elif data[x-i][y-i] == 'L':
            i = num_rows + num_cols
        i += 1
    
    i = 1
    while x - i >= 0:
        if data[x-i][y] == '#':
            counter += 1
            i = num_rows + num_cols
        elif data[x-i][y] == 'L':
            i = num_rows + num_cols
        i += 1
        
    i = 1
    while x - i >= 0 and y + i <= num_cols - 1:
        if data[x-i][y+i] == '#':
            counter += 1
            i = num_rows + num_cols
        elif data[x-i][y+i] == 'L':
            i = num_rows + num_cols
        i += 1
    
    i = 1
    while y - i >= 0:
        if data[x][y-i] == '#':
            counter += 1
            i = num_rows + num_cols
        elif data[x][y-i] == 'L':
            i = num_rows + num_cols
        i += 1
    
    i = 1
    while y + i <= num_cols - 1:
        if data[x][y+i] == '#':
            counter += 1
            i = num_rows + num_cols
        elif data[x][y+i] == 'L':
            i = num_rows + num_cols
        i += 1
    
    i = 1
    while x + i <= num_rows - 1 and y - i >=0:
        if data[x+i][y-i] == '#':
            counter += 1
            i = num_rows + num_cols
        elif data[x+i][y-i] == 'L':
            i = num_rows + num_cols
        i += 1
    
    i = 1
    while x + i <= num_rows - 1:
        if data[x+i][y] == '#':
            counter += 1
            i = num_rows + num_cols
        elif data[x+i][y] == 'L':
            i = num_rows + num_cols
        i += 1
    
    i = 1
    while x + i <= num_rows - 1 and y + i <= num_cols - 1:
        if data[x+i][y+i] == '#':
            counter += 1
            i = num_rows + num_cols
        elif data[x+i][y+i] == 'L':
            i = num_rows + num_cols
        i += 1
        
    if counter >= 5:
        return True
    return False
```

```python
def get_occupied_seats2(data):

    repeat = True
    
    num_rows = len(data)
    num_cols = len(data[0])
    
    iterations = 0
    
    while repeat:
        
        iterations += 1
        data_copy = data.copy()
        
        for x in range(num_rows):
            for y in range(num_cols):
                
                if y < num_cols - 1:
                    if data_copy[x][y] == 'L':
                        if check_to_fill2(data_copy, x, y):
                            data[x] = data[x][:y] + '#' + data[x][y+1:]
                    elif data_copy[x][y] == '#':
                        if check_to_vacate2(data_copy, x, y):
                            data[x] = data[x][:y] + 'L' + data[x][y+1:]
                else:
                    if data_copy[x][y] == 'L':
                        if check_to_fill2(data_copy, x, y):
                            data[x] = data[x][:y] + '#'
                    elif data_copy[x][y] == '#':
                        if check_to_vacate2(data_copy, x, y):
                            data[x] = data[x][:y] + 'L'
                
        if data_copy == data:
            repeat = False
    
    seat_count = 0
    
    for row in data:
        seat_count += row.count('#')
    
    return seat_count
```
